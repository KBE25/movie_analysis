### Loading IMDB Data

```python
import sqlite3
conn = sqlite3.Connection("Data/im.db")
```

```python
movie_basics_table = """
SELECT *
FROM movie_basics;
"""
movie_basics = pd.read_sql_query(movie_basics_table, conn)
movie_basics.head()
```

```python
movie_basics.info()
```

```python
movie_basics.describe(include="all")
```

```python
movie_ratings_table = """
SELECT *
FROM movie_ratings;
"""
movie_ratings = pd.read_sql_query(movie_ratings_table, conn)
movie_ratings.head()
```

```python
movie_ratings.info()
```

```python
movie_ratings.describe(include="all")
```

```python
movie_akas_table = """
SELECT *
FROM movie_akas;
"""
movie_akas = pd.read_sql_query(movie_akas_table, conn)
movie_akas.head()
```

```python
movie_principals_table = """
SELECT *
FROM principals;
"""
movie_principals = pd.read_sql_query(movie_principals_table, conn)
movie_principals.head()
```

```python
movie_principals.info()
```

```python
movie_persons_table = """
SELECT *
FROM persons;
"""
movie_persons = pd.read_sql_query(movie_persons_table, conn)
movie_persons.head()
```

```python
movie_persons.info()
```

```python
movie_directors_table = """
SELECT mb.*, mr.averagerating, mr.numvotes, pe.primary_name, pe.birth_year, pe.death_year, pe.primary_profession, pr.category
From movie_basics as mb
JOIN movie_ratings as mr
USING(movie_id)
JOIN principals as pr
USING(movie_id)
JOIN persons as pe
USING(person_id)
WHERE pr.category='director'
ORDER BY numvotes;
"""
movie_directors = pd.read_sql_query(movie_directors_table, conn)
movie_directors.head()
```

```python
movie_directors.info()
```

```python
movie_writers_table = """
SELECT mb.*, mr.averagerating, mr.numvotes, pe.primary_name, pe.birth_year, pe.death_year, pe.primary_profession, pr.category
From movie_basics as mb
JOIN movie_ratings as mr
USING(movie_id)
JOIN principals as pr
USING(movie_id)
JOIN persons as pe
USING(person_id)
JOIN directors as dr
USING(person_id)
WHERE pr.category='writer';
"""

movie_writers = pd.read_sql_query(movie_writers_table, conn)
movie_writers.head()
```

```python
movie_writers.info()
```

```python
movie_actors_table = """
SELECT mb.*, mr.averagerating, mr.numvotes, pe.primary_name, pe.birth_year, pe.death_year, pe.primary_profession, pr.category
From movie_basics as mb
JOIN movie_ratings as mr
USING(movie_id)
JOIN principals as pr
USING(movie_id)
JOIN persons as pe
USING(person_id)
JOIN directors as dr
USING(person_id)
WHERE pr.category='actor';
"""

movie_actor = pd.read_sql_query(movie_actors_table, conn)
movie_actor.head()
```

```python
movie_actor.info()
```

```python
movie_actress_table = """
SELECT mb.*, mr.averagerating, mr.numvotes, pe.primary_name, pe.birth_year, pe.death_year, pe.primary_profession, pr.category
From movie_basics as mb
JOIN movie_ratings as mr
USING(movie_id)
JOIN principals as pr
USING(movie_id)
JOIN persons as pe
USING(person_id)
JOIN directors as dr
USING(person_id)
WHERE pr.category='actress'
ORDER BY averagerating DESC;
"""

movie_actress = pd.read_sql_query(movie_actress_table, conn)
movie_actress.head()
```
